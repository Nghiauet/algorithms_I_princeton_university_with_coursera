```python
class MagicNode:
    def __init__(self,data) -> None:
        self.data = data 
        self.next = None 
        self.prev = None 
```

```python
class Dequeue:
    def __init__(self) :
        self.first = self.last = None
        self.size = 0 
    def isEmpty(self):
        if self.first ==None:
            return True
        return False
    
    def size(self):
        return self.size()
    def addFirst(self,item):
        temp = MagicNode(item)
        if self.first == None:
            self.first = self.last = temp 
        else:
            temp.prev = self.first
            self.first.next = temp 
            self.first = temp
    def addLast(self, item):
        temp = MagicNode(item)
        if self.last == None:
            self.last = self.first = temp 
        else:
            self.last.prev = temp
            temp.next = self.last 
            self.last = temp 
    def removeFirst(self):
        if self.isEmpty():
            print('NoSuchElementException for removeFirst')
        else:
            if self.first.prev == None:
                temp = self.first
                self.first = self.last = None
                print('removefirst pop: ',temp.data)
                return temp.data
            else:
                temp = self.first.prev 
                temp.next = None 
                # save data before shift to left
                data =self.first.data  
                self.first = temp
                print('removefirst pop: ',data)

                return data 
    def removeLast(self):
        if self.isEmpty():
            print('NoSuchElementException for removeFirst')
        else:
            if self.last.next == None:
                temp = self.last 
                self.last = self.first = None
                print("removelast pop: ", temp.data)
                return temp.data
            else: 
                temp = self.last.next 
                temp.prev = None 
                # save data before shift to right
                data = self.last.data 
                self.last = temp 
                print('removeLast pop: ',data)
                return data 
    def iterator(self):
        temp = self.last
        print("forward", end = ": ") 
        while temp is not None:
            print(temp.data, end = " ")
            temp = temp.next
        print()
    def rev_iterator(self):
        temp = self.first
        print("backward", end = ": ") 
        while temp is not None:
            print(temp.data, end =" ")
            temp = temp.prev
        print()
```

```python
class DynamicStack:
    def __init__(self):
        self._n = 0
        self.capacity = 1
        # init with 1 element
        self.stack = [0] * 1
        
    def isEmpty(self):
        if self._n == 0 : 
            return True
        else: return False
        
    def _resize(self, new_capacity):
        assert new_capacity >= self._n , " capacity must always > n"
        new_array = [None] * new_capacity
        for i in range(self._n):
            new_array[i] = self.stack[i]
        self.stack = new_array
        self.capacity = new_capacity
        
    def push(self, item):
        if self._n + 1 > self.capacity:
            self._resize((self.capacity+1) * 2)
        self._n  += 1
        # print("size and capacity: ", self._n, self.capacity)
        self.stack[self._n-1] = item
        
    def pop(self):
        if self.isEmpty():
            print("pop an empty stack")
        if self._n <= self.capacity//4: 
            self._resize(self.capacity // 2)
        item = self.stack[self._n-1]
        self.stack[self._n-1] = None
        self._n -= 1 
        print("size and capacity: ", self._n, self.capacity)
        print("item pop: ", item)
        return item
    
    def iterator(self):
        print("iterator")
        for i in range(self._n):
            print(self.stack[i], end = " ")
        print()
```

```python
import random
class RandomizedQueue:
    def __init__(self):
        self._n = 0
        self.capacity = 1
        # init with 1 element
        self.stack = [0] * 1
        
    def isEmpty(self):
        if self._n == 0 : 
            return True
        else: return False
        
    def _resize(self, new_capacity):
        assert new_capacity >= self._n , " capacity must always > n"
        new_array = [None] * new_capacity
        for i in range(self._n):
            new_array[i] = self.stack[i]
        self.stack = new_array
        self.capacity = new_capacity
        
    def push(self, item):
        if self._n + 1 > self.capacity:
            self._resize((self.capacity+1) * 2)
        self._n  += 1
        # print("size and capacity: ", self._n, self.capacity)
        self.stack[self._n-1] = item
        
    def pop(self):
        if self.isEmpty():
            print("pop an empty stack")
        
        if self._n <= self.capacity//4: 
            self._resize(self.capacity // 2)
            
        # take rundom index
        index = random.randint(0,self._n-1)
        item = self.stack[index] 
        # swap with last element
        self.stack[index] = self.stack[self._n-1]
        
        self.stack[self._n-1] = None
        self._n -= 1 
        # print("size and capacity: ", self._n, self.capacity)
        # print("item pop: ", item)
        return item
    
    def iterator(self):
        # pop till it empty
        print("begin to iterator")
        while not self.isEmpty():
            print(self.pop(), end = " ")
        print()
        print("end iterator")
```
