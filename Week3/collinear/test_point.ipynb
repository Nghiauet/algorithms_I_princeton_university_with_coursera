```python
import matplotlib.pyplot as plt

# Create a list of points
points = [(3,7), (6,4), (1,1), (4,3), (7,3)]
# Extract x and y coordinates from points
x = [p[0] for p in points]
y = [p[1] for p in points]
# connect the points with lines
lines = []
for i in range(len(points)):
    for j in range(i+1, len(points)):
        if points[i][0] != points[j][0] and points[i][1] != points[j][1]:
            lines.append([points[i][0], points[i][1], points[j][0], points[j][1]])

# plot the points in the square grid
plt.scatter(x, y)
# connect the points with lines have more than 3 points

for i in range(len(lines)):
    plt.plot([lines[i][0], lines[i][2]], [lines[i][1], lines[i][3]], 'r-')


# Show the plot
# grid on
plt.grid(True)
# range of x-axis and y-axis
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.show()
```
